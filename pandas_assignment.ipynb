```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data0 = pd.read_csv("states0.csv")
data1 = pd.read_csv("states1.csv")
data2 = pd.read_csv("states2.csv")
data3 = pd.read_csv("states3.csv")
data4 = pd.read_csv("states4.csv")
data5 = pd.read_csv("states5.csv")
data6 = pd.read_csv("states6.csv")
data7 = pd.read_csv("states7.csv")
data8 = pd.read_csv("states8.csv")
data9 = pd.read_csv("states9.csv")
```

# 2

```python
datalist = [data0, data1, data2, data3, data4, data5, data6, data7, data8, data9]
data = pd.concat(datalist,ignore_index=True)
data = data.drop(data.columns[0],axis=1)
data
```

# 5

```python
data['Income'] = data.Income.str[1:]
data['Income'] = pd.to_numeric(data.Income)
data.head()
```

# 6

```python
 gender=data.GenderPop.str.split("_")
data["Men"] = gender.str.get(0)
data["Women"]= gender.str.get(1)
```

```python
data.head()
```

# 7

```python
data["Men"] = data.Men.str[:-1]
data["Women"] = data.Women.str[:-1]
```

```python
data["Men"] = pd.to_numeric(data.Men)
data["Women"] = pd.to_numeric(data.Women)
data.head()
```

```python
data = data.fillna(value={"Women": data.TotalPop - data.Men})
```

# 14

```python
Hispanic = data["Hispanic"]
Hispanic.fillna(0)
Hispanic.drop_duplicates()
Hispanic = Hispanic.str.extract('(\d+)').astype(int)

White = data["White"]
White.fillna(0)
White.drop_duplicates()
White = White.str.extract('(\d+)').astype(int)

Black = data["Black"]
Black.fillna(0)
Black.drop_duplicates()
Black = Black.str.extract('(\d+)').astype(int)

Native = data["Native"]
Native.fillna(0)
Native.drop_duplicates()
Native = Native.str.extract('(\d+)').astype(int)

Asian = data["Asian"]
Asian.fillna(0)
Asian.drop_duplicates()
Asian = Asian.str.extract('(\d+)').astype(int)
```
